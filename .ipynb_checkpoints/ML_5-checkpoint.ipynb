```python
# problem come dealing with text data instead of numeric data so we deal with that using one hot encoding
```

```python
import pandas as pd
df = pd.read_csv('homeprices.csv')
df
```

```python
#first step create variable dummy coulmn
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis=1)
merged.drop(['town','west windsor'],axis=1,inplace=True)
merged
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = merged[['area','monroe township','robinsville']]
Y = merged['price']
```

```python
model.fit(X,Y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,Y)
```

```python
## using sklearn onehotencoder
df
## do one label encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
## the change
dfle = df
le.fit_transform(dfle.town)
```

```python
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['area','town']]
Y = dfle.price
model2 = LinearRegression()
model2.fit(X,Y)
```

```python
model2.predict([[2500,0]])
```

```python
model2.score(X,Y)
```

```python
## exerise solve
import pandas as pd
df = pd.read_csv('Exersise 4.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])

merged = pd.concat([df,dummies],axis=1)
merged
```

```python
merged.drop(['Car Model','Audi A5'],axis=1,inplace=True)
merged
```

```python
model3 = LinearRegression()
X = merged[['Mileage','Age(yrs)','Mercedez Benz C class','BMW X5']]
Y = merged['Sell Price($)']
model3.fit(X,Y)
```

```python
model3.predict([[45000,4,0,1]])
```

```python
model3.score(X,Y)
```
